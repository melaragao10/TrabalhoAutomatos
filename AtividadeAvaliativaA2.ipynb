```python
#Melissa Aragão Leite - 33999554
#Exercício 1

# ----------------------------------------------------
# Autômato finito para L = (a|b|c)*
#
# Definição formal do AFD:
# Q = {q0, qerr}
# Σ = {a, b, c}
# q0 = estado inicial
# F = {q0}
# δ (função de transição):
#   δ(q0, a) = q0
#   δ(q0, b) = q0
#   δ(q0, c) = q0
#   δ(q0, outro) = qerr
#   δ(qerr, x) = qerr (estado de erro)
#
# Ou seja: o autômato aceita qualquer cadeia (inclusive vazia)
# formada apenas pelos símbolos a, b e c.
# ----------------------------------------------------

# Verifica se uma cadeia pertence à linguagem do Exercício 1
def pertence_linguagem_1(cadeia: str) -> bool:
    """
    Qualquer cadeia formada apenas pelos símbolos 'a', 'b' e 'c',
    incluindo a cadeia vazia (""), é aceita.

    Parâmetro:
        cadeia (str): string digitada pelo usuário
    Retorno:
        bool: True se pertence à linguagem, False caso contrário
    """

    # Define o alfabeto válido da linguagem
    alfabeto = {"a", "b", "c"}

    # all() percorre cada símbolo e só retorna True
    # se TODOS estiverem dentro do alfabeto
    return all(simbolo in alfabeto for simbolo in cadeia)

while True:
    # Pede ao usuário para digitar uma cadeia
    cadeia = input("Digite uma cadeia (ou 'sair' para encerrar): ")

    # Se o usuário digitar "sair", o programa acaba
    if cadeia.lower() == "sair":
        print("Encerrando o programa. Até logo!")
        break

    # Chama a função que verifica se a cadeia pertence à linguagem
    if pertence_linguagem_1(cadeia):
        print(f"A cadeia '{cadeia}' pertence à linguagem ✅")
    else:
        print(f"A cadeia '{cadeia}' NÃO pertence à linguagem ❌")
```

```python
#Artur Almeida - 1433363650
# Exercício 6

# Módulo: Validador de Cadeias (Comprimento >= 3)
#alunos:artur,Jhonathan,melissa,Gabriel,Alejandro
# Baseado no Exercício 6: Cadeias com comprimento maior ou igual a 3
# Alfabeto: {a, b, c}

# 1. Representação Formal do Autômato Finito Determinístico (AFD)
# Q = {q0, q1, q2, q3}  # Conjunto de Estados
# F = {'q3'}             # Conjunto de Estados Finais (Aceitação)

ESTADOS_FINAIS = {'q3'}
ESTADO_INICIAL = 'q0'
ALFABETO = {'a', 'b', 'c'}

# Função de Transição Delta (d)
# { (estado_atual, simbolo): proximo_estado, ... }
TRANSICOES = {
    # q0 (0 caracteres lidos) -> q1
    ('q0', 'a'): 'q1', ('q0', 'b'): 'q1', ('q0', 'c'): 'q1',
    # q1 (1 caractere lido) -> q2
    ('q1', 'a'): 'q2', ('q1', 'b'): 'q2', ('q1', 'c'): 'q2',
    # q2 (2 caracteres lidos) -> q3 (ACEITAÇÃO)
    ('q2', 'a'): 'q3', ('q2', 'b'): 'q3', ('q2', 'c'): 'q3',
    # q3 (3+ caracteres lidos) -> q3 (laço de aceitação)
    ('q3', 'a'): 'q3', ('q3', 'b'): 'q3', ('q3', 'c'): 'q3',
}

# 2. Função de Simulação do AFD

def validar_cadeia(cadeia):

    estado_atual = ESTADO_INICIAL

    if not cadeia:
        # Cadeia vazia é rejeitada, pois 0 < 3.
        return ("REJEITADA", ESTADO_INICIAL)

    print(f"\n| Simulação para '{cadeia}'")
    print(f"| Estado Inicial: {estado_atual}")

    for simbolo in cadeia:
        # 1. Verifica se o símbolo pertence ao alfabeto
        if simbolo not in ALFABETO:
            return (f"REJEITADA (Símbolo '{simbolo}' inválido)", estado_atual)

        # 2. Executa a transição
        try:
            proximo_estado = TRANSICOES[(estado_atual, simbolo)]
            print(f"| Lendo '{simbolo}': {estado_atual} -> {proximo_estado}")
            estado_atual = proximo_estado
        except KeyError:
            # Caso a tabela esteja mal definida ou haja erro de lógica.
            return (f"REJEITADA (Transição não definida)", estado_atual)

    # 3. Verifica o estado final
    if estado_atual in ESTADOS_FINAIS:
        return ("ACEITA", estado_atual)
    else:
        return ("REJEITADA", estado_atual)

# Módulo Interativo para o Usuário


print("Alfabeto aceito: {a, b, c}")
print("Digite 'sair' para encerrar a aplicação.")

while True:
    try:
        cadeia_usuario = input("\nInsira a cadeia para validar: ").strip().lower()

        if cadeia_usuario in ('sair'):
            print("Encerrando o validador. Até a próxima!")
            break

        if not cadeia_usuario:
            print("Por favor, insira uma cadeia de caracteres.")
            continue

        # Chama a função de validação
        resultado, estado_final = validar_cadeia(cadeia_usuario)

        print(f"| RESULTADO: {resultado}")
        print(f"| Estado Final: {estado_final}")

    except Exception as e:
        print(f"\nOcorreu um erro inesperado: {e}")
        break
```

```python
#Alejandro Lopes - 34125701
#Exercício 3
def validador_comprimento_3(cadeia: str) -> str:
    """
    Esta função só deixa passar cadeias com exatamente 3 letras.
    """
    #variável marcar progresso.
    # Começa no 'q0', o  ponto de partida
    estado_atual = 'q0'

    for letra in cadeia:

        #a letra tem que ser 'a', 'b' ou 'c'.
        if letra not in ('a', 'b', 'c'):

            return f"A cadeia '{cadeia}' é REJEITADA (só pode ter 'a', 'b' ou 'c')."

        # A cada letra válida avança
        if estado_atual == 'q0':
            estado_atual = 'q1'

        elif estado_atual == 'q1':
            estado_atual = 'q2'

        elif estado_atual == 'q2':
            estado_atual = 'q3'

        elif estado_atual == 'q3':
            estado_atual = 'q_erro'
            break

    #Depois de olhar todas as letras confere onde parou
    # A cadeia só é válida se tiver parado no 3
    if estado_atual == 'q3':
        return f"A cadeia '{cadeia}' é ACEITA."
    else:
        # Se parar antes ou cair em erro rejeita.
        return f"A cadeia '{cadeia}' é REJEITADA."

# --- Demonstração do Validador ---
print("### Validador para Cadeias de Comprimento 3 ###")

# Testando com um texto de 3 letras
print(validador_comprimento_3("bca"))

# Testando com um texto menor
print(validador_comprimento_3("ab"))

# Testando com um texto maior
print(validador_comprimento_3("abca"))

print("-" * 20)

# --- Teste Interativo ---
cadeia_usuario = input("Digite uma cadeia para validar: ")
print(validador_comprimento_3(cadeia_usuario))
```

```python
# Atividade 4 - AFD – Cadeias de comprimento diferente de 3 (Σ = {a, b, c})
# Autor: Gabriel Henrique Luiz Mendes (RGBM: 33890897)
# Observação: Código pronto para Colab / Python 3.10+
# ============================================================

from typing import Dict, Set, Iterable, Optional
from dataclasses import dataclass

# ------------------------------------------------------------
# 1) Definições e utilidades
# ------------------------------------------------------------

@dataclass(frozen=True)
class DFA:
    """
    Representa um Autômato Finito Determinístico (AFD).

    Definição formal:
      M = (Q, Σ, δ, q0, F)
        - Q: conjunto de estados
        - Σ: alfabeto
        - δ: função de transição δ: Q × Σ → Q
        - q0: estado inicial
        - F: conjunto de estados de aceitação

    Esta classe apenas guarda os componentes; a simulação
    fica por conta da classe DFASimulator.
    """
    states: Set[str]
    alphabet: Set[str]
    delta: Dict[str, Dict[str, str]]
    start: str
    accept: Set[str]


class DFASimulator:
    """
    Simulador de AFD com checagens básicas.

    Uso típico:
        dfa = construir_afd_len_diferente_de_3({'a','b','c'})
        sim = DFASimulator(dfa)
        sim.validate("abc")  # -> False (rejeita, pois |w|=3)
    """
    def _init_(self, dfa: DFA, *, strict: bool = True):
        """
        Parâmetros:
          - dfa: instância de DFA (com δ idealmente total).
          - strict: se True, rejeita imediatamente ao ver símbolo
                    fora de Σ ou transição ausente.
                    se False, ignora símbolo fora de Σ (não recomendado).
        """
        self.dfa = dfa
        self.strict = strict
        self._validar_forma_basica()

    def _validar_forma_basica(self) -> None:
        """Checagens de integridade simples do AFD."""
        assert self.dfa.start in self.dfa.states, "q0 não pertence a Q."
        assert self.dfa.accept.issubset(self.dfa.states), "F ⊄ Q."
        # Verificação fraca da δ (aceita δ parcial, mas sinaliza).
        for q, trans in self.dfa.delta.items():
            assert q in self.dfa.states, f"Estado desconhecido em δ: {q}"
            for a, qnext in trans.items():
                assert a in self.dfa.alphabet, f"Símbolo '{a}' não pertence a Σ."
                assert qnext in self.dfa.states, f"Transição aponta para estado inexistente: {qnext}"

    def validate(self, w: str) -> bool:
        """
        Processa a cadeia w e retorna True se for aceita,
        ou False caso contrário.
        """
        q = self.dfa.start
        for a in w:
            if a not in self.dfa.alphabet:
                if self.strict:
                    # Símbolo inválido => rejeita
                    return False
                else:
                    # Modo tolerante (não recomendado para aulas):
                    # apenas ignora o símbolo fora de Σ
                    continue
            # δ total assumida para este exercício. Se não houver,
            # considere tratar exceção ou rejeitar.
            if q not in self.dfa.delta or a not in self.dfa.delta[q]:
                # Transição ausente => rejeita (δ parcial)
                return False
            q = self.dfa.delta[q][a]
        return q in self.dfa.accept


def imprimir_tabela_transicao(dfa: DFA, order_states: Optional[Iterable[str]] = None) -> None:
    """
    Imprime uma tabela de transição legível no console.
    """
    states = list(order_states) if order_states is not None else sorted(dfa.states)
    alphabet = sorted(dfa.alphabet)
    # Cabeçalho
    header = ["Estado"] + alphabet
    largura = [max(6, len(h)) for h in header]
    def fmt_row(cells):
        return " | ".join(str(c).ljust(w) for c, w in zip(cells, largura))

    print(fmt_row(header))
    print("-" * (sum(largura) + 3*(len(largura)-1)))

    for q in states:
        row = [q]
        for a in alphabet:
            qnext = dfa.delta.get(q, {}).get(a, "—")
            row.append(qnext)
        print(fmt_row(row))


# ------------------------------------------------------------
# 2) Construção do AFD para |w| != 3
# ------------------------------------------------------------

def construir_afd_len_diferente_de_3(alphabet: Set[str]) -> DFA:
    """
    Constrói um AFD que reconhece L = { w ∈ Σ* | |w| ≠ 3 }.

    Ideia: "contar" o comprimento até 4.
      - q0: já li 0 símbolos
      - q1: já li 1 símbolo
      - q2: já li 2 símbolos
      - q3: já li 3 símbolos     (ÚNICO ESTADO NÃO-ACEITADOR)
      - q4: já li ≥ 4 símbolos    (estado de saturação)

    Transições (para todo x ∈ Σ):
      δ(q0,x)=q1, δ(q1,x)=q2, δ(q2,x)=q3, δ(q3,x)=q4, δ(q4,x)=q4

    Estados de aceitação: F = {q0, q1, q2, q4}
    """
    states = {"q0", "q1", "q2", "q3", "q4"}
    start = "q0"
    accept = {"q0", "q1", "q2", "q4"}

    # Como a transição não depende do símbolo, mapeamos todos para o mesmo próximo estado.
    delta = {
        "q0": {s: "q1" for s in alphabet},
        "q1": {s: "q2" for s in alphabet},
        "q2": {s: "q3" for s in alphabet},
        "q3": {s: "q4" for s in alphabet},
        "q4": {s: "q4" for s in alphabet},
    }

    return DFA(states=states, alphabet=alphabet, delta=delta, start=start, accept=accept)


# ------------------------------------------------------------
# 3) Demonstração e testes rápidos
# ------------------------------------------------------------

def rodar_testes_basicos(sim: DFASimulator) -> None:
    """
    Executa alguns testes de sanidade e imprime resultados.
    """
    print("\n=== Testes básicos ===")
    casos = [
        "",        # |w|=0 -> aceita
        "a",       # |w|=1 -> aceita
        "bc",      # |w|=2 -> aceita
        "abc",     # |w|=3 -> rejeita
        "bbbc",    # |w|=4 -> aceita
        "cabac",   # |w|=5 -> aceita
        "bbb",     # |w|=3 -> rejeita
        "bbcc",    # |w|=4 -> aceita
        "c",       # |w|=1 -> aceita
        "aXc",     # contém símbolo fora de Σ -> rejeita (strict=True)
    ]
    for w in casos:
        print(f"{w!r:>8} -> {'ACEITA' if sim.validate(w) else 'REJEITA'}")

    # Asserts para dar confiança (não param a execução no Colab se der tudo certo)
    assert sim.validate("") is True
    assert sim.validate("a") is True
    assert sim.validate("bc") is True
    assert sim.validate("abc") is False
    assert sim.validate("bbb") is False
    assert sim.validate("bbcc") is True


# ------------------------------------------------------------
# 4) Modo interativo (opcional)
# ------------------------------------------------------------

def modo_interativo(sim: DFASimulator) -> None:
    """
    Permite ao usuário digitar cadeias e ver ACEITA/REJEITA.
    Digite 'sair' para encerrar.
    """
    print("\n=== Modo interativo ===")
    print("Digite cadeias sobre Σ = {a,b,c}. Digite 'sair' para terminar.")
    while True:
        w = input("cadeia> ").strip()
        if w.lower() == "sair":
            print("Encerrando modo interativo.")
            break
        ok = sim.validate(w)
        print("Resultado:", "ACEITA ✅" if ok else "REJEITA ❌")


# ------------------------------------------------------------
# 5) Execução principal (Colab/Notebook)
# ------------------------------------------------------------

if _name_ == "_main_":
    # Alfabeto do exercício
    sigma = {"a", "b", "c"}

    # Constrói o AFD e o simulador
    dfa = construir_afd_len_diferente_de_3(sigma)
    sim = DFASimulator(dfa, strict=True)

    print("AFD para L = { w ∈ Σ* | |w| ≠ 3 } com Σ = {a,b,c}")
    print("Estados:", dfa.states)
    print("Estado inicial:", dfa.start)
    print("Estados de aceitação:", dfa.accept)
    print("\nTabela de transição:")
    imprimir_tabela_transicao(dfa, order_states=["q0","q1","q2","q3","q4"])

    # Testes rápidos
    rodar_testes_basicos(sim)

    # Para usar o modo interativo, descomente a linha abaixo no Colab:
    # modo_interativo(sim)
```

```python
# ----------------------------------------------------
# DEFINIÇÃO FORMAL DO AFD (Exercício 17)
# Linguagem: Cadeias em {a,b,c}* que iniciam com 'a' e terminam com 'c'.
# ----------------------------------------------------
ESTADO_INICIAL = 'Q0'
ESTADOS_FINAIS = {'Q2'}
ALFABETO = {'a', 'b', 'c'}

# Tabela de Transição (Delta[Estado Atual][Símbolo] = Próximo Estado)
TRANSICOES = {
    'Q0': {'a': 'Q1', 'b': 'Q3', 'c': 'Q3'},
    'Q1': {'a': 'Q1', 'b': 'Q1', 'c': 'Q2'}, # Q1: Começou com 'a', mas não termina em 'c'
    'Q2': {'a': 'Q1', 'b': 'Q1', 'c': 'Q2'}, # Q2: Estado Final (Começou com 'a' E terminou em 'c')
    'Q3': {'a': 'Q3', 'b': 'Q3', 'c': 'Q3'}  # Q3: Estado Morto/Rejeição
}

def validar_cadeia(cadeia):
    """Simula o AFD para validar a cadeia de entrada passo a passo."""

    # 1. Tratamento da cadeia vazia (epsilon)
    if not cadeia:
        return ESTADO_INICIAL in ESTADOS_FINAIS # Retorna False (Q0 não é final)

    estado_atual = ESTADO_INICIAL
    print(f"\n--- Simulação para '{cadeia}' ---")
    print(f"Estado inicial: {estado_atual}")

    # 2. Iteração e Simulação
    for i, simbolo in enumerate(cadeia):
        # Validação do Alfabeto
        if simbolo not in ALFABETO:
            print(f"ERRO: Símbolo '{simbolo}' na posição {i+1} não pertence ao alfabeto. REJEITADA.")
            return False

        # Aplica a Função de Transição Delta(q, x) = q'
        proximo_estado = TRANSICOES[estado_atual][simbolo]
        print(f"   Lendo '{simbolo}': {estado_atual} -> {proximo_estado}")
        estado_atual = proximo_estado

    # 3. Verificação do Estado Final
    print(f"\nFim da leitura. Estado final alcançado: {estado_atual}")

    if estado_atual in ESTADOS_FINAIS:
        return True
    else:
        return False

# ----------------------------------------------------
# SEÇÃO DE INTERAÇÃO COM O USUÁRIO (PARA APRESENTAÇÃO)
# ----------------------------------------------------

print("="*60)
print("AFD Validador: Cadeias que iniciam com 'a' e terminam com 'c'")
print(f"Alfabeto: {ALFABETO}")
print("Digite 'sair' a qualquer momento para encerrar.")
print("="*60)

while True:
    entrada_usuario = input("\nDigite a cadeia para validação: ").strip()

    if entrada_usuario.lower() == 'sair':
        print("\nSimulação encerrada.")
        break

    if not entrada_usuario:
        print("Entrada vazia. Tente novamente ou digite 'sair'.")
        continue

    # Chama a função de validação com a entrada do usuário
    resultado_usuario = validar_cadeia(entrada_usuario)

    # Exibe o resultado final
    print("\nRESULTADO FINAL:")
    if resultado_usuario:
        print(f"A cadeia '{entrada_usuario}' é **ACEITA** pela linguagem.")
    else:
        print(f"A cadeia '{entrada_usuario}' é **REJEITADA** pela linguagem.")
```
